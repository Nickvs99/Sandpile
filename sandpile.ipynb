```python
import numpy as np

from sandpile_model import Sandpile_model
from plots import plot_time_series, plot_size_probability
```

```python
grid_sizes = np.linspace(10, 100, num=5, dtype=int)

size_probability_dataset = []
labels = []

for grid_size in grid_sizes:
    model = Sandpile_model(grid_size=grid_size, n_steps=10000, crit_values=[2, 4], n_grain_types=2)
    model.run()
    
    size_probability_dataset.append(model.collect_size_probability_data(n_bins=15))
    labels.append(f"Size: {grid_size}")
```

```python
plot_size_probability(size_probability_dataset, labels=labels)
```
